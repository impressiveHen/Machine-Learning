```python
# load data
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
all_data, all_labels = load_wine(True)
# 178 data points, 13 dimensions
print(all_data.shape)
# 178 class labels 1,2,3
print(all_labels.shape)
data = all_data[0:(59+71)]
label = all_labels[0:(59+71)]
print(data.shape)
print(label.shape)
```

```python

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def cross_entropy_loss_accuracy(x,y,w,b):
    m = x.shape[0]
    z = np.dot(x,w.T) + b
    a = sigmoid(z)
    loss = (np.dot(-y,np.log(a)) - np.dot((1-y),np.log(1-a))) / m
    predictions = np.zeros((m,1))
    predictions[a>=0.5] = 1
    accuracy = np.sum(predictions==y[:,None])/m*100
    return loss, accuracy
```

```python

# sklearn logistic regression gradient descent 
from sklearn.linear_model import LogisticRegression
 
num_iter = 3000
grad_iter_loss = np.zeros((num_iter,1))
grad_iter_accuracy = np.zeros((num_iter,1))
for i in range(num_iter):
    # c is regularization constant 
    model = LogisticRegression( C=1e10, solver='sag', max_iter=i)
    model.fit(data, label)
    w = model.coef_
    b = model.intercept_
    loss, accuracy = cross_entropy_loss_accuracy(data,label,w,b)
    grad_iter_loss[i] = loss
    grad_iter_accuracy[i] = accuracy

    
    
    
    
    
```

```python
print(model.get_params())
```

```python
# plot loss, iteration
plt.figure(figsize = (8,6))
plt.plot(np.arange(num_iter), grad_iter_loss)
plt.xlabel('number of iteration')
plt.ylabel('loss')
plt.title('Loss of Logistic Regression Gradient Descent')
plt.show()
```

```python

# plot accuracy, iteration
plt.figure(figsize = (8,6))
plt.plot(np.arange(num_iter), grad_iter_accuracy)
plt.xlabel('number of iteration')
plt.ylabel('accuracy%')
plt.title('accuracy of Logistic Regression Gradient Descent')
plt.show()
```

```python
'''
>>> a
array([[1, 2, 3],
       [2, 3, 4],
       [3, 4, 5]])
>>> np.insert(a,0,1,axis = 1)
array([[1, 1, 2, 3],
       [1, 2, 3, 4],
       [1, 3, 4, 5]])
'''
import random as rand

# define class logistic regression coordinate descent
class log_reg_coord_des:
    '''
    binary classification
    activation function: sigmoid
    loss function: cross entropy
    
    '''
    def __init__(self, lr=0.001, op_meth = 'gradient', max_iter=100):
        self.learning_rate = lr
        self.optimize_method = op_meth
        self.max_iteration = max_iter
        self.coef_ = None
        self.x = None
        self.y = None
    
    def fit(self,x,y):
        
        self.x = np.insert(x,0,1,axis=1)
        self.y = y[:,None]
        self.coef_ = np.zeros((1,self.x.shape[1]))
        for i in range(self.max_iteration):
            z = np.dot(self.x,self.coef_.T)
            a = self.sigmoid(z)
            if self.optimize_method == 'gradient':
                self__grad_update_weights(a)
            elif self.optimize_method == 'coordinate':
                self.__coord_update_weights(a)
            elif self.optimize_method == "random":
                self.__rand_update_weights(a)
            else:    
                pass
            
    def sigmoid(self,z):
        return 1 / (1 + np.exp(-z))
    
    def __grad_update_weights(self,a):
        dw = np.sum((a-self.y)*self.x,axis=1)/self.x.shape[0]
        self.coef_ =  self.coef_ - self.learning_rate*(a-self.y)*self.x
        
    def __coord_update_weights(self,a):
        dw = np.sum((a-self.y)*self.x,axis=0)/self.x.shape[0]
        dim_i = np.argmax(np.abs(dw))
        self.coef_[0,dim_i] = self.coef_[0,dim_i] - self.learning_rate*dw[dim_i]
    
    def __rand_update_weights(self,a):
        dw = np.sum((a-self.y)*self.x,axis=0)/self.x.shape[0]
        dim_i = rand.randint(0,13)
        self.coef_[0,dim_i] = self.coef_[0,dim_i] - self.learning_rate*dw[dim_i]
    
```

```python

def cross_entropy_loss_accuracy_coord(x,y,w):
    m = x.shape[0]
    z = np.dot(x,w.T) 
    a = sigmoid(z)
    loss = (np.dot(-y,np.log(a)) - np.dot((1-y),np.log(1-a))) / m
    predictions = np.zeros((m,1))
    predictions[a>=0.5] = 1
    accuracy = np.sum(predictions==y[:,None])/m*100
    return loss, accuracy

```

```python
num_iter = 3000
coord_iter_loss = np.zeros((num_iter,1))
coord_iter_accuracy = np.zeros((num_iter,1))
for i in range(num_iter):
    # c is regularization constant 
    model = log_reg_coord_des(lr=1e-5, op_meth='coordinate', max_iter=i)
    model.fit(data, label)
    w = model.coef_
    loss, accuracy = cross_entropy_loss_accuracy_coord(np.insert(data,0,1,axis=1),label,w)
    coord_iter_loss[i] = loss
    coord_iter_accuracy[i] = accuracy
```

```python
# plot loss, iteration

plt.figure(figsize = (8,6))
plt.plot(np.arange(num_iter), coord_iter_loss)
plt.xlabel('number of iteration')
plt.ylabel('loss')
plt.title('Loss of Logistic Regression coordinate Descent')
plt.show()
```

```python

# plot accuracy, iteration
plt.figure(figsize = (8,6))
plt.plot(np.arange(num_iter), coord_iter_accuracy)
plt.xlabel('number of iteration')
plt.ylabel('accuracy%')
plt.title('accuracy of Logistic Regression coordinate Descent')
plt.show()
```

```python
num_iter = 3000
rand_iter_loss = np.zeros((num_iter,1))
rand_iter_accuracy = np.zeros((num_iter,1))
for i in range(num_iter):
    # c is regularization constant 
    model = log_reg_coord_des(lr=1e-5, op_meth='random', max_iter=i)
    model.fit(data, label)
    w = model.coef_
    loss, accuracy = cross_entropy_loss_accuracy_coord(np.insert(data,0,1,axis=1),label,w)
    rand_iter_loss[i] = loss
    rand_iter_accuracy[i] = accuracy
```

```python
# plot loss, iteration

plt.figure(figsize = (8,6))
plt.plot(np.arange(num_iter), rand_iter_loss)
plt.xlabel('number of iteration')
plt.ylabel('loss')
plt.title('Loss of Logistic Regression random Descent')
plt.show()
```

```python
# plot accuracy, iteration
plt.figure(figsize = (8,6))
plt.plot(np.arange(num_iter), rand_iter_accuracy)
plt.xlabel('number of iteration')
plt.ylabel('accuracy%')
plt.title('accuracy of Logistic Regression random Descent')
plt.show()
```
